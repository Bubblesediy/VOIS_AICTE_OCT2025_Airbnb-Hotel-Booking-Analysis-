```python
# Cheking Hotel Type

# Cheking the count of unique hotel type present
unique_hotel_type = aba_df['hotel'].value_counts()
print("There are",aba_df['hotel'].nunique(),"Unique Hotel Type present in Dataset,each accompanied by its respective count :")
print(unique_hotel_type)
print("---------------------------------------------------------- ")

# Checking Average(mean) adr for each hotel type
hotel_type_avg_adr = aba_df.groupby('hotel').describe()['adr']['mean'].round(2)
print("Average adr of each Hotel type present :")
print(hotel_type_avg_adr)
print("------------------------------------------------------------")

# Checking Average(mean) number of stays in week nights for each hotel type
hotel_type_avg_week_nights=aba_df.groupby('hotel').describe()['stays_in_week_nights']['mean'].astype(int)
print("Average Number of Stays in Week Nights for each hotel type present :")
print(hotel_type_avg_week_nights)
print("------------------------------------------------------------")

# Checking Average(mean) number of stays in weekend nights for each hotel type
hotel_type_avg_weekend_nights = aba_df.groupby('hotel').describe()['stays_in_weekend_nights']['mean'].astype(int)
print("Average Number of Stays in Weekend Nights for each hotel type present :")
print(hotel_type_avg_weekend_nights)
print("------------------------------------------------------------")

# Max total of special requests received by each hotel_type
hotel_type_max_special_requests = aba_df.groupby('hotel').describe()['total_of_special_requests']['max']
print("Maximum Number of Total of Special Requests received by each hotel type")
print(hotel_type_max_special_requests)
print("------------------------------------------------------------")

# Average lead time per hotel type
print("Average Lead Time for each hotel type :")
hotel_type_avg_lead_time = aba_df.groupby('hotel').describe()['lead_time']['mean'].round(2)
print(hotel_type_avg_lead_time)
```

```python
# Monthly Analysis based on arrival date

# Examining how the Number of Bookings varies each Month based on the arrival dates to understand any recurring trends.

# All months Bookings count for each hotel type
booking_counts_by_hotel_type = aba_df.groupby(['hotel', 'arrival_date_month']).size().reset_index(name='booking_count')
print("Count of Bookings made for each month across all hotel types.")
print(booking_counts_by_hotel_type)
print('--------------------------------------------------------------')

# Top 3 months with the maximum booking counts for each hotel type
booking_counts_by_hotel_type_month = aba_df.groupby(['hotel', 'arrival_date_month']).size()
top_3_months_by_hotel_type_max = booking_counts_by_hotel_type_month.groupby('hotel', group_keys=False).nlargest(3).reset_index(name='booking_count')
print("Top 3 months for each hotel type when the maximum number of bookings occurred :")
print(top_3_months_by_hotel_type_max)
print('----------------------------------------------------------')

# Top 3 months with the minimum booking counts for each hotel type
top_3_months_by_hotel_type_min = booking_counts_by_hotel_type_month.groupby('hotel', group_keys=False).nsmallest(3).reset_index(name='booking_count')
print("Top 3 months for each hotel type when the minimum number of bookings occurred :")
print(top_3_months_by_hotel_type_min)
print('-----------------------------------------------------------')

# Analysing based on Each hotel type

# Means in all hotel type which hotel type got maximum booking count along with respective month of max booking
# Top 3 Maximum Booking counts across all hotel types.
booking_counts_by_hotel_type_in_month = aba_df.groupby(['hotel', 'arrival_date_month']).size().nlargest(3).reset_index(name='booking_count')
print("Top 3 Maximum Booking counts across all hotel types and months")
print(booking_counts_by_hotel_type_in_month)
print('--------------------------------------------------------------')

# Maximum Booking count of each hotel type across all months.
booking_counts_by_each_hotel_type_in_month = aba_df.groupby(['hotel', 'arrival_date_month']).size().reset_index(name='booking_count')
top3_by_hotel_type_in_month = booking_counts_by_each_hotel_type_in_month.groupby(['hotel'], group_keys=False).apply(lambda x: x.nlargest(1, 'booking_count')).reset_index(drop=True)

print("Months in which Maximum Booking is made for each hotel type")
print(top3_by_hotel_type_in_month)
```

```python
# Yearly Analysis based on last review

# Convert 'last_review' to datetime
aba_df['last_review'] = pd.to_datetime(aba_df['last_review'])

# Extract Year and Month from 'last_review'
aba_df['year'] = aba_df['last_review'].dt.year.astype(float).astype('Int64')


# All years' Booking count for each room type
booking_counts_by_room_type_yearly = aba_df.groupby(['room_type', 'year']).size().reset_index(name='booking_count_yearly')
print("Count of Bookings made for each year across all room types.")
print(booking_counts_by_room_type_yearly)
print('--------------------------------------------------------------')

# Top 3 years with the maximum booking counts for each room type
booking_counts_by_rtype_yearly = aba_df.groupby(['room_type', 'year']).size()
top_3_years_by_room_type_max = booking_counts_by_rtype_yearly.groupby('room_type', group_keys=False).nlargest(3).reset_index(name='booking_count_yearly')
print("Top 3 years for each room type when the maximum number of bookings occurred :")
print(top_3_years_by_room_type_max)
print('----------------------------------------------------------')

# Top 3 years with the minimum booking counts for each room type
top_3_years_by_room_type_min = booking_counts_by_rtype_yearly.groupby('room_type', group_keys=False).nsmallest(3).reset_index(name='booking_count_yearly')
print("Top 3 years for each room type when the minimum number of bookings occurred :")
print(top_3_years_by_room_type_min)
print('-----------------------------------------------------------')

# Analysing based on Each neighbourhood group

# Means in all neighbourhood_group which room type got maximum booking count along with respective year of max booking
# Top 3 Maximum Booking counts across all neighbourhood groups and room_type.
booking_counts_by_room_type_in_ngroup_yearly = aba_df.groupby(['neighbourhood_group', 'room_type', 'year']).size().nlargest(3).reset_index(name='booking_count_yearly')
print("Top 3 Most Booked room types across all neighbourhood groups over all years")
print(booking_counts_by_room_type_in_ngroup_yearly)
print('--------------------------------------------------------------')

# Maximum Booking count of each room type across all neighbourhood groups.
booking_counts_by_each_room_type_in_ngrp_yearly = aba_df.groupby(['neighbourhood_group', 'room_type', 'year']).size().reset_index(name='booking_count_yearly')
top3_by_room_type_in_ngrp_yearly = booking_counts_by_each_room_type_in_ngrp_yearly.groupby(['neighbourhood_group', 'room_type'], group_keys=False).apply(lambda x: x.nlargest(1, 'booking_count_yearly')).reset_index(drop=True)

print("Year in which Maximum Booking is made for each room type across all neighbourhood groups")
print(top3_by_room_type_in_ngrp_yearly)
```

```python
# Month with Maximum Bookings for each year in each hotel type

# Convert 'arrival_date_year', 'arrival_date_month', and 'arrival_date_day_of_month' to datetime
aba_df['arrival_date'] = pd.to_datetime(aba_df['arrival_date_year'].astype(str) + '-' + aba_df['arrival_date_month'] + '-' + aba_df['arrival_date_day_of_month'].astype(str), errors='coerce')

# Filter out rows with NaT in 'arrival_date'
aba_df.dropna(subset=['arrival_date'], inplace=True)

# Extract Year and Month from 'arrival_date'
aba_df['year'] = aba_df['arrival_date'].dt.year
aba_df['month'] = aba_df['arrival_date'].dt.month


# Group by hotel, year, and month, count occurrences
booking_counts_by_month_yearly = aba_df.groupby(['hotel','year', 'month']).size().reset_index(name='booking_count_monthly')

# Find the month with the maximum bookings for each year per hotel type
max_booking_month_by_hotel_year = booking_counts_by_month_yearly.loc[booking_counts_by_month_yearly.groupby(['hotel','year'])['booking_count_monthly'].idxmax()]

print("Month with Maximum Bookings for each year in each hotel type:")
print(max_booking_month_by_hotel_year)
```

```python
# Distribution Analysis

# 25%, 50%, and 75% Quartile Distribution of ADR Across Hotel Types
adr_summary = aba_df.groupby('hotel')['adr'].describe(percentiles=[0.25, 0.5, 0.75])

# Additional Observation: Interquartile Range (IQR)
adr_summary['iqr'] = adr_summary['75%'] - adr_summary['25%']

print("Quartile Distribution of ADR Across Hotel Types with Interquartile Range (IQR):")
print(adr_summary.loc[:, ['25%', '50%', '75%', 'iqr']])
print('-------------------------------------------------------------------------')

# Hotel Type Distribution Summary
hotel_summary = aba_df.groupby('hotel')['hotel'].describe()
print("Hotel Type Distribution Summary")
print(hotel_summary)
```

#### Chart - 3: Average Price Trend Over Years (Univariate)


```python
# Convert 'arrival_date' to datetime and extract the year
aba_df['arrival_date'] = pd.to_datetime(aba_df['arrival_date'])
aba_df['year'] = aba_df['arrival_date'].dt.year

# Group by year and calculate the average adr
average_adr_by_year = aba_df.groupby('year')['adr'].mean()

# Plot the trend
plt.figure(figsize=(10, 6))
plt.plot(average_adr_by_year.index, average_adr_by_year.values, marker='o')
plt.title('Average ADR Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Average ADR')
plt.grid(True)
plt.show()

print("Average ADR per Year")
print(average_adr_by_year)
```

#### Chart - 4: Average Price Trend Over Years by Room Type (Bivariate)

```python
# Convert 'arrival_date' to datetime
aba_df['arrival_date'] = pd.to_datetime(aba_df['arrival_date'])

# Extract Year from 'arrival_date'
aba_df['year'] = aba_df['arrival_date'].dt.year

# Group by year and hotel, calculate the average adr
average_adr_by_year_hotel = aba_df.groupby(['year', 'hotel'])['adr'].mean().unstack()

# Plot the trend for each hotel type over years
plt.figure(figsize=(12, 8))
for hotel_type in average_adr_by_year_hotel.columns:
    plt.plot(average_adr_by_year_hotel.index, average_adr_by_year_hotel[hotel_type], marker='o', label=hotel_type)

plt.title('Average ADR Trend Over Years by Hotel Type')
plt.xlabel('Year')
plt.ylabel('Average ADR')
plt.legend()
plt.grid(True)

plt.show()
```

#### Chart - 5: Average Price Trend Over Months by Room Type (Bivariate)

```python
# predicting trends (monthly) between avg price & room type

# Convert 'arrival_date' to datetime
aba_df['arrival_date'] = pd.to_datetime(aba_df['arrival_date'])

# Create a pivot table to calculate the mean adr for each hotel type and each month
monthly_avg_adr_by_hotel = pd.pivot_table(aba_df, values='adr', index=aba_df['arrival_date'].dt.month, columns='hotel', aggfunc='mean')

# Plot the monthly average adr for each hotel type
plt.figure(figsize=(12,8))
monthly_avg_adr_by_hotel.plot()
plt.title('Monthly Average ADR by Hotel Type')
plt.xlabel('Month')
plt.ylabel('Average ADR')
plt.legend(title='Hotel Type', loc='upper left')

plt.show()
```

#### Chart - 6: Average Price by Room Type (Bivariate)

```python
# Calculate the average adr for each hotel type
average_adr = aba_df.groupby('hotel')['adr'].mean().reset_index()

# Create the bar plot for average adr
plt.figure(figsize=(8, 6))  # Creating a new figure
sns.barplot(x='hotel', y='adr', data=average_adr)
plt.xlabel('Hotel Type')
plt.ylabel('Average ADR')
plt.title('Average ADR by Hotel Type')

plt.show()
```
